```python
# The ability to provide statistical data for categorical variables

def cat_stats(col):
    cat_df = bike.groupby(col)['count'].agg(['sum', 'mean','count']).sort_values('sum',ascending = False)
    cat_df['sum_perc']=cat_df['sum']/bike['count'].sum()*100
    cat_df['count_perc']=cat_df['count']/bike['count'].count()*100
    return round(cat_df,2)
```

```python
#Plotting functions for categorical variables

def cat_plots(col,x,y):
    plt.figure(figsize = (x,y))
    plt.subplot(1,2,1)
    sns.barplot(x=col, y='count', data= bike)
    plt.subplot(1,2,2)
    sns.barplot(x=col, y='count', data= bike, hue = 'year', palette = 'Paired')
    plt.legend(labels=['2018', '2019'])
    return
```

## Developing Manual Models using Statsmodel

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculateVIF(df):
    # Create a DataFrame to store feature names and their respective VIFs
    vif = pd.DataFrame()
    vif['Features'] = df.columns
    # Calculate VIF for each feature
    vif['VIF'] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    # Round VIF values to two decimal places
    vif['VIF'] = round(vif['VIF'], 2)
    # Sort DataFrame by VIF values in descending order
    vif = vif.sort_values(by="VIF", ascending=False)
    return vif

# Ensure that the DataFrame contains only numeric data types and handle any missing or infinite values
```
